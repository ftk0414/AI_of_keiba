```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
```

```python
#二次元正規分布の確率密度を返す関数
def gaussian(x,mu):
    #分散共分散行列の行列式
    det = np.linalg.det(sigma)
    #print(det)
    #分散共分散行列の逆行列
    inv = np.linalg.inv(sigma)
    n = x.ndim
    #print(inv)
    return np.exp(-np.diag((x - mu)@inv@(x - mu).T)/2.0) / (np.sqrt((2 * np.pi) ** n * det))

d = 0.00001
m = 0.01
t = 100

def kakusan(x,p,d,t,m):
    inv = np.linalg.inv(sigma)
    return (m/(4 * np.pi * d)**1.5) * np.exp(-np.diag((x - p)@inv@(x - p).T)/(4*d*t))

sigma = np.array([[1,0],[0,1]])
```

```python
#関数に投入するデータを作成
x = y = np.arange(-0.5, 0.5, 0.01)
X, Y = np.meshgrid(x, y)
z = np.c_[X.ravel(),Y.ravel()]

#2変数の平均値を指定
p00 = np.array([0,0])
p10 = np.array([0.125,0])
p20 = np.array([-0.125,0])

p04 = np.array([0,0.04])
p14 = np.array([0.125,0.04])
p24 = np.array([-0.125,0.04])

p08 = np.array([0,0.08])
p18 = np.array([0.125,0.08])
p28 = np.array([-0.125,0.08])

p04_ = np.array([0,-0.04])
p14_ = np.array([0.125,-0.04])
p24_ = np.array([-0.125,-0.04])

p08_ = np.array([0,-0.08])
p18_ = np.array([0.125,-0.08])
p28_ = np.array([-0.125,-0.08])


Z00 = kakusan(z,p00,d,t,m)
Z10 = kakusan(z,p10,d,t,m)
Z20 = kakusan(z,p20,d,t,m)

Z04 = kakusan(z,p04,d,t,m)
Z14 = kakusan(z,p14,d,t,m)
Z24 = kakusan(z,p24,d,t,m)

Z08 = kakusan(z,p08,d,t,m)
Z18 = kakusan(z,p18,d,t,m)
Z28 = kakusan(z,p28,d,t,m)

Z04_ = kakusan(z,p04_,d,t,m)
Z14_ = kakusan(z,p14_,d,t,m)
Z24_ = kakusan(z,p24_,d,t,m)

Z08_ = kakusan(z,p08_,d,t,m)
Z18_ = kakusan(z,p18_,d,t,m)
Z28_ = kakusan(z,p28_,d,t,m)


Z = Z00 + Z10 +Z20 + Z04 + Z14 +Z24 +Z08 + Z18 +Z28 + Z04_ + Z14_ +Z24_ +Z08_ + Z18_ +Z28_
shape = X.shape
Z = Z.reshape(shape)

#二次元正規分布をplot
fig = plt.figure(figsize = (15, 15))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm)

plt.show()
```
